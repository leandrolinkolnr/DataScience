## Criando e excluindo variaveis/columns no dataSet

```python
import pandas as pd
```

```python
dados = pd.read_csv('./Arquivos/aluguel_residencial.csv', sep = ';')
dados.head()
```

### Criando nova coluna com o valor bruto
#### Valor + Condominio + IPTU

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head()
```

### Criando coluna 'Valor por m²'

```python
dados['Valor por m²'] = (dados['Valor'] / dados['Area'])
dados.head()
```

#### Arredondando

```python
dados['Valor por m²'] = dados['Valor por m²'].round(2) # Arredondando
dados.head()
```

## Variavel tipo agregado

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados.head(10)
```

# Excluindo variaveis ou colunas

### Metodo del

```python
# Excluindo dos dados acima
del dados['Valor Bruto']
dados.head()
```

### Usando os dados criados acima para criar dataSet auxiliar

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado','Valor por m²']])
dados_aux.head()
```

### Metodo Pop

```python
dados_aux.pop('Valor por m²')
dados_aux.head()
```

##### Criando o dataSet novamente para poder excluir

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado','Valor por m²']])
dados_aux.head(2)
```

### Metodo Drop

```python
# Serve para linhas e colunas, entao precisamos passar o eixo
dados_aux.drop(['Valor por m²', 'Tipo Agregado'], axis = 1, inplace = True)
dados_aux.head()
```

### Exportando os dados para o csv

```python
dados.to_csv('./Arquivos/aluguel_residencial.csv', sep=';', index=False)
```
