## Criando novas variaveis no dataSet

```python
import pandas as pd
```

```python
dados = pd.read_csv('./Arquivos/aluguel_residencial.csv', sep = ';')
dados.head()
```

### Criando nova coluna com o valor bruto
#### Valor + Condominio + IPTU

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head()
```

### Criando coluna 'Valor por m²'

```python
dados['Valor por m²'] = (dados['Valor'] / dados['Area'])
dados.head()
```

#### Arredondando

```python
dados['Valor por m²'] = dados['Valor por m²'].round(2) # Arredondando
dados.head()
```

## Variavel tipo agregado

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados.head(10)
```
