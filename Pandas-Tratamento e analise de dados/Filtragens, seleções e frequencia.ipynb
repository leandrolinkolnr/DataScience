# Filtragens e Frequencia

```python
import pandas as pd
```

```python
dados = pd.read_csv("./Arquivos/aluguel_residencial.csv", sep = ';')
dados.head()
```

### Selecionando apenas imoveis classificados como Apartamento

```python
selecaoBoolean = dados['Tipo'] == 'Apartamento'  # Aqui ele marca cada um como True ou false
selecaoBoolean.head()
```

```python
selecao = dados[selecaoBoolean]    # Aqui ele pega os dados que sao marcados como true
selecao.head()
```

```python
# Aqui temos a frequencia
selecao.shape[0]
```

### Selecionando apenas imoveis classificados como "Casa", "casa de condominio" ou "casa de vila"

```python
 # Aqui ele marca como True ou false
selecaoBoolean = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
selecaoBoolean.head()
```

```python
selecao = dados[selecaoBoolean]    # Aqui ele pega os dados que sao marcados como true
selecao.head(11)
```

```python
# Aqui temos a frequencia
selecao.shape[0]
```

### Selecionando imoveis apenas com area entre 60 e 100 metros

```python
selecaoBoolean = (dados['Area'] >= 60) & (dados['Area'] <= 100)
selecaoBoolean.head()
```

```python
selecao = dados[selecaoBoolean]    # Aqui ele pega os dados que sao marcados como true
selecao.head(11)
```

```python
# Aqui temos a frequencia
selecao.shape[0]
```

### Selecionando imoveis apenas com 4 quartos e aluguel menos que R$ 2000

```python
selecaoBoolean = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
selecaoBoolean.head()
```

```python
selecao = dados[selecaoBoolean]    # Aqui ele pega os dados que sao marcados como true
selecao.head(11)
```

```python
# Aqui temos a frequencia
selecao.shape[0]
```
