```python
import pandas as pd
```

```python
dados = pd.read_csv("Arquivos/aluguel_residencial.csv", sep =';')
dados.head()    #Ja notamos valores null
```

### Função isnull:  verifica onde há numeros faltantes
###### Marcando como true ou false

```python
dados.isnull()
```

### Função notnull:  verifica onde há numeros faltantes
###### Ao contrario do isnull

```python
dados.notnull()
```

### Info: Função que nos permite ver quantos dados não null existem em cada coluna

```python
dados.info()
```

### Com essa função, pegamos todos os itens do dataSet que possuem valor null na coluna "Valor"

```python
dados[dados['Valor'].isnull()]
```

### Dropna: Apagar os itens que tem valores null

```python
# Apagar dados que possuem null na coluna valor
dados.dropna(subset = ['Valor'], inplace= True)
```

```python
#Verificando se os valores null foram apagados
dados[dados['Valor'].isnull()]
```

# Tratamentos diferentes para outras variaveis

#### Não podemos apagar todos os dados que tem condominio null, pois casa não paga condominio

### Verificando os dados que possuem condominio como null

```python
dados[dados['Condominio'].isnull()]
```

### Vamos selecionar openas os apartamentos que tem condomio null

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
dados[selecao].head()
```

### Vamos eliminar todos os itens que foram selecionados (apartamento com condominio null)

```python
dados = dados[~selecao]
dados.head()
```

### Verificando se os dados foram eliminados

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
dados[selecao]
```

### Verificando os dados restantes que tem condominio null

```python
dados[dados['Condominio'].isnull()].head()
```

## fillna: Atribuir o valor 0 a todos que tem condominio null

```python
dadosteste = dados.fillna(0)
```

### Verificando que todos os valores null de todas as colunas foram colocados = 0

```python
dadosteste[dadosteste['Condominio'].isnull()].head()
```

```python
dadosteste[dadosteste['IPTU'].isnull()].head()
```

## Definindo explicitamente como substituir os valores null em cada coluna

```python
# Igual a dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
dados.fillna({'Condominio': 0, 'IPTU': 0}, inplace = True)
dados.head()
```

## Verificando

```python
dados[dados['Condominio'].isnull()]
```

```python
dados[dados['IPTU'].isnull()]
```

```python
dados.info()
```

## Exportando os novos dados após limpeza
#### Substituindo o arquivo anterior

```python
dados.to_csv("Arquivos/aluguel_residencial.csv", sep =';', index = False)
```
