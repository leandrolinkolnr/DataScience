### Importando a base
- Base: 'BaseInadimplencia.xlsx'
    - Aba: 'teste'

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Importando o perceptron
from sklearn.linear_model import Perceptron
clf = Perceptron(tol=1e-3, random_state=0)
```

```python
# sheet_name = o arquivo tem outras abas e eu quero a aba treino
base = pd.read_excel("BaseInadimplencia.xlsx",sheet_name='treino')
base.head(2)
```

## Criando o modelo usando uma reta que separe esses pontos


### Formula: 

#### y = (- clf.intercept_ - clf.coef_[0][0].x)/clf.coef_[0][1]

```python
# Selecionando o X e o y
x = base[['SaldoConta','SaldoInvestimento']]
y = base.Situacao
```

```python
# Fazendo o fit com o modelo
clf.fit(x, y)
```

#### Criando um scatter plot usando as informações da base e com as cores sendo a classificação do modelo

```python
# c = cor pra separar de acordo com a coluna situação)

# Aqui estamos fazendo um gráfico de dispersão
fig, ax = plt.subplots()
ax.scatter(base['SaldoConta'], base['SaldoInvestimento'], c=base['Situacao'])


# Aqui estamos fazendo um gráfico de linha do Perceptron para separar os dados
x = np.arange(-7, 7)
y = (-clf.intercept_-clf.coef_[0][0]*x)/clf.coef_[0][1]
ax.plot(x,y,c='r')


plt.show()
```
