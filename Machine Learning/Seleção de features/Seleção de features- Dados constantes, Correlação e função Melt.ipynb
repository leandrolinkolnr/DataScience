# Função Melt: Reorganizando as colunas para que fiquem em uma linha.

#### Ela derreta as colunas em linhas, ou seja, ela diminui as colunas do dataSet, levando as informações paa as linhas
#### Util quando temos muitas dimensões

id_vars = Quais colunas que você deseja manter intactas

var_name = Nome da nova coluna que terá as colunas derretidas

value_name = Nome da nova coluna que armazenará os valores das colunas derretidas.

```python
dados_plot = pd.melt(dados_plot, id_vars="diagnostico",
                    var_name="exames",
                    value_name='valores')
dados_plot.head()
```

# Função para plotar o graficos somente dos exames que queremos

```python
def grafico_violino(valores, inicio, fim):
    
    dados_plot = pd.concat([diagnostico, valores.iloc[:,inicio:fim]],axis = 1)
    dados_plot = pd.melt(dados_plot, id_vars="diagnostico",
                         var_name="exames",
                         value_name='valores')
    plt.figure(figsize=(9, 9))
    sns.violinplot(x = "exames", y = "valores", hue = "diagnostico",
                   data = dados_plot, split= True)
    plt.xticks(rotation = 90)
```
