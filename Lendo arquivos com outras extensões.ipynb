# Lendo arquivos de diferentes extensões

## Visualizando arquivo Json

```python
import pandas as pd
```

```python
json = open("Arquivos/aluguel.json")
print(json.read())
```

## Transformando o Json em um dataSet

```python
df_json = pd.read_json("Arquivos/aluguel.json")
df_json
```

## Visualizando arquivo Txt

```python
txt = open("Arquivos/aluguel.txt")
print(txt.read())
```

## Transformando o txt em um dataSet

```python
df_txt = pd.read_table("Arquivos/aluguel.txt")
df_txt
```

## Transformando o xlsx em um dataSet

```python
df_xlsx = pd.read_excel("Arquivos/aluguel.xlsx")
df_xlsx
```

## Transformando o html em um dataSet

```python
df_html = pd.read_html("Arquivos/dados_html_1.html")
df_html
#df_html[0]      # Mostrar como DataFrame
```

## Caso der erro 403 forbidden

```python
import requests

url = "https://www.federalreserve.gov/releases/h3/current/default.htm"

#Informações para fingir ser um navegador
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

#juntamos tudo com a requests
r = requests.get(url, headers=header)
```

```python
#E finalmente usamos a função read_html do pandas

dfs = pd.read_html(r.text)
```
